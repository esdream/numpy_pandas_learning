Matplotlib Tutorial
---

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 50)
y1 = 2 * x + 1
y2 = x ** 2
# 处于plt.gigure()下方的plot会绘制在一张图上，直至下一个figure出现，切换至下一张图
plt.figure(num=-1, figsize=(2, 3))
plt.plot(x, y1)

# 函数绘图，label参数是legend上的标签
plt.figure(num=3, figsize=(8, 5))
l1, = plt.plot(x, y2, label='up')
l2, = plt.plot(x, y1, color='gray', linewidth=3.0, linestyle='--', label='down')
# 绘制图例,handles是legend中要添加的图像对象（例如这里为l1，其中在定义时必须加上一个逗号），labels参数表示图中显示
# 的图像label名，如果和图像对象的label不一样，则显示plt.legend()中labels参数传入的label名
plt.legend(handles=[l1,], labels=['aaa',], loc='best')

# 横纵坐标的取值范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))

# 坐标轴描述
plt.xlabel('x label')
plt.ylabel('y label')

# ticks是坐标轴上的小标
new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks)
# 如果传入两个参数，第一个是坐标数值，第二个是数据对应的标签
# 使用r'$word$'可以转换图中的字体，空格、特殊符号等需要加反斜杠输入
plt.yticks([-2, -1.8, -1, 1.22, 4,],
          [r'$really\ bad$', r'$bad\ \alpha$', r'$normal$', r'$good$', r'$really\ good$'])

# 修改图变宽
# gca = 'get current axis'，matplotlib中的边框以轴表示
ax = plt.gca()
# ax.spines获取对应位置的轴
# ax.spines['right'].set_color('none')
# ax.spines['top'].set_color('none')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# 设置x轴和y轴
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# 设置轴的位置
ax.spines['bottom'].set_position(('data', -1))
ax.spines['left'].set_position(('data', 0))

plt.show()
```
