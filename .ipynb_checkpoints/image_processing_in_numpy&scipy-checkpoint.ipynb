```python
import numpy as np
from scipy import misc
from glob import glob
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# misc中的face方法是一个图片样张
f = misc.face()
```

```python
misc.imsave('./image/face.png', f)
face = misc.imread('face.png')
# 创建raw文件
face.tofile('./image/face.raw')
face_from_raw = np.fromfile('./image/face.raw', dtype=np.uint8)
# reshape raw 文件
face_from_raw.shape = (768, 1024, 3)
```

```python
# 对于大数据，使用np.memmap做内存映射。数据从文件中读取，没有加载到内存
face_memmap = np.memmap('./image/face.raw', dtype=np.uint8, shape=(768, 1024, 3))
```

```python
for i in range(10):
    im = np.random.randint(0, 255, 10000).reshape((100, 100))
    misc.imsave('./image/random_%02d.png' % i, im)

filelist = glob('./image/random*.png')
filelist.sort()
```

```python
# 读取成灰度图像
f_gray = misc.face(gray=True)
```

```python
# 使用matplotlib和imshow在matplotlib图形内部显示图像
plt.imshow(f_gray, cmap=plt.cm.gray)
# 设置最小值和最大值增加对比度
plt.imshow(f_gray, cmap=plt.cm.gray, vmin=30, vmax=200)
```

```python
plt.imshow(f_gray, cmap=plt.cm.gray, vmin=30, vmax=200)
# 删除坐标轴和刻度
plt.axis('off')
# 绘制轮廓线，输入必须为2D矩阵（灰度图像）
plt.contour(f_gray, [50, 200])
```

```python
# 放大局部图像
plt.imshow(f_gray[320:340, 510:530], cmap=plt.cm.gray)
```

```python
# 修改部分像素值
f_gray[100:120] = 255
lx, ly = f_gray.shape
# np.ogrid方法会创建一个N维（N为ogrid中参数的个数，例如这里是2维）对象。并返回在N个维坐标轴上的投影（例如这里X为0-1024, Y为0-768）
X, Y = np.ogrid[0:lx, 0:ly]
# 掩膜
mask = (X - lx / 2) ** 2 + (Y - ly / 2) ** 2 > lx * ly / 4
f_gray[mask] = 0
# 象征索引
f_gray[range(400), range(400)] = 255
plt.imshow(f_gray, cmap=plt.cm.gray)
```
