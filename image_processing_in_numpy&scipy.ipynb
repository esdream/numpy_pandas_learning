```python
import numpy as np
from scipy import misc
from glob import glob
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# misc中的face方法是一个图片样张
f = misc.face()
misc.imsave('./image/face.png', f)

face = misc.imread('face.png')
# 创建raw文件
face.tofile('./image/face.raw')
face_from_raw = np.fromfile('./image/face.raw', dtype=np.uint8)
# reshape raw 文件
face_from_raw.shape = (768, 1024, 3)
```

```python
# 对于大数据，使用np.memmap做内存映射。数据从文件中读取，没有加载到内存
face_memmap = np.memmap('./image/face.raw', dtype=np.uint8, shape=(768, 1024, 3))
```

```python
for i in range(10):
    im = np.random.randint(0, 255, 10000).reshape((100, 100))
    misc.imsave('./image/random_%02d.png' % i, im)

filelist = glob('./image/random*.png')
filelist.sort()
```

```python
# 使用matplotlib和imshow在matplotlib图形内部显示图像
f = misc.face(gray=True)
plt.imshow(f, cmap=plt.cm.gray)
# 设置最小值和最大值增加对比度
plt.imshow(f, cmap=plt.cm.gray, vmin=30, vmax=200)
```
